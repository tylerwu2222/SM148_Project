```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
messy = pd.read_csv('time_series_covid19_confirmed_US.csv')
messy.drop(messy.iloc[:, :5], inplace = True,axis=1)
messy.drop(messy.iloc[:, 2:6], inplace = True,axis=1)
messy.head()
```

```python
tidy = pd.melt(messy,id_vars = ['Admin2','Province_State'],var_name = 'date',value_name='cases')
tidy.head()
tidy.to_csv('time_series_covid19_confirmed_tidy.csv',index=False)
```

```python
Autauga = tidy.loc[tidy.Admin2 == 'Autauga']
plt.plot(Autauga.date,Autauga.cases)
```
