### Scatterplots for Polarity vs COVID stats

```python
def plot_scatter(x,y):
    x_range = (cases_df[x].min(), cases_df[x].max())
    y_range = (cases_df[y].min(), cases_df[y].max())
    print("x_range: {0} y_range: {1}".format(x_range,y_range))
    cvs = ds.Canvas(x_range=x_range, y_range=y_range, plot_height=300, plot_width=300)
    agg = ds.Canvas().points(cases_df, x, y)
    return ds.tf.set_background(ds.tf.shade(agg, cmap=cc.fire), "black")
```
