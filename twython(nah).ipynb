```python
geocode = '34.0522,-118.2437,15mi' # 15 mile radius of LA
search_results = twitter.search(q='covid',count=50, geocode=geocode).get('statuses') # max count = 100
data = []
for tweet in search_results:
    # print(tweet)
    data_row = []
    try:
        data_row.append(tweet.get('user').get('name'))
        data_row.append(tweet.get('text'))
        data_row.append(tweet.get('user').get('location'))
        data_row.append(tweet.get('created_at'))
    except TwythonError as e:
        print(e)
    data.append(data_row)
df = pd.DataFrame(data,columns = ['username','text','location','created_at'])
df
```
